```python
# Function for GOOGLE CLOUD STORAGE!
def upload_to_bucket(blob_name, local_path):
    """
    Uploads a file to the bucket.

    Parameters
    ------------
    - blob_name (str):
        The name of the blob in the bucket.
    - local_path (str):
        The local path to the file to upload.
    """
    blob = bucket.blob(blob_name)
    blob.upload_from_filename(local_path)

def bucket_to_local(blob_name, local_path):
    """
    Downloads a blob from the bucket to a local file.

    Parameters
    ------------
    - blob_name (str):
        The name of the blob in the bucket.
    - local_path (str):
        The local path to save the downloaded file.
    """
    blob = bucket.blob(blob_name)
    blob.download_to_filename(local_path)

def get_all_blobs():
    """
    Returns a list of all blob names in the bucket.
    """
    filename=list(bucket.list_blobs())
    files = [blob.name for blob in filename]
    return files

def rename_blobs(original_blob, renamed_blob):
    """
    Renames a blob in the bucket.
    
    Parameters
    ------------
    - original_blob (str):
        The name of the blob to rename.
    - renamed_blob (str):
        The new name for the blob.
    """
    if original_blob in get_all_blobs():
        blob = bucket.blob(original_blob)
        new_blob = bucket.rename_blob(blob, renamed_blob)
        print(f"Blob {original_blob} renamed to {renamed_blob}")
    else:
        print(f"Blob {original_blob} does not exist.")

def rename_blob_folder(original_folder, renamed_folder):
    """
    Renames all blobs in a folder in the bucket.

    Parameters
    ------------
    - original_folder (str):
        The name of the folder containing blobs to rename.
    - renamed_folder (str):
        The new name for the folder.
    """
    original_folder = f'{original_folder}/'
    renamed_folder = f'{renamed_folder}/'
    blobs = [i for i in get_all_blobs() if i.startswith(original_folder)]
    for old_blob in blobs:
        new_blob = old_blob.replace(original_folder, renamed_folder)
        old_blob = bucket.blob(old_blob)
        a = bucket.rename_blob(old_blob, new_blob)
    print("Rename success!")

def bucket_to_local_folder(blob_folder, local_folder, ZIP = False):
    """
    Downloads all blobs in a folder from the bucket to a local folder.

    Parameters
    ------------
    - blob_folder (str):
        The name of the folder in the bucket containing blobs to download.
    - local_folder (str):
        The local path to save the downloaded blobs.
    - ZIP (bool):
        If True, the downloaded folder will be zipped.
    """
    os.makedirs(local_folder, exist_ok=True)
    blob_folder = f'{blob_folder}/'
    blobs = [i for i in get_all_blobs() if i.startswith(blob_folder)]
    for blob in blobs:
        local_path = blob.replace(blob_folder, f'{local_folder}/')
        bucket_to_local(blob, local_path)
    if ZIP:
        shutil.make_archive(local_folder, 'zip', local_folder)

def count_class(src, returnType: Literal["count", "full"]="count") -> pd.DataFrame:
    """
    Counts the number of instances in each class from the given source directory.
    The source directory should contain subdirectories for each class, and each subdirectory should contain the images belonging to that class.

    Parameters
    ------------
    - src: 
        The source directory containing the class subdirectories.
    - returnType:
        The type of return value. If "count", it returns a list of counts for each class. 
        If "full", it returns a DataFrame with class names and their respective counts.
    """
    class_counts = []
    class_names = []

    # Get the list of class directories
    for class_name in os.listdir(src):
        class_path = os.path.join(src, class_name)
        if os.path.isdir(class_path):
            count = len(os.listdir(class_path))
            class_counts.append(count)
            class_names.append(class_name)
    
    if returnType == "count":
        return class_counts
    elif returnType == "full":
        df = pd.DataFrame({'Class': class_names, 'len': class_counts})
        df['len'] = df['len'].astype(int)
        return df

# No longer in use, in pursue of strartified k-fold
def split_train_eval(dst_name, composition = [0.85, 0.15], seed=rd.randint(0, 100000)):
    """
    Splits the dataset into training and evaluation sets based on the specified composition. 

    The dataset is expected to be organized in a directory structure where each class has its own subdirectory.

    DIR WILL BE CREATED IN DATA/!TEMP/<dst_name>

    Parameters
    ------------
    - dst_name (str):
        The name of the destination directory where the split datasets will be saved.
    - composition (list):
        A list containing two float values that represent the proportion of the training and evaluation sets, respectively.
    """
    temp = {}
    assert sum(composition) == 1, "Composition must sum to 1"
    
    # Split dataset into train and val equally on each class
    for class_name in tqdm(os.listdir(f'DATA/!FINAL/train_val'), desc="Splitting data into train and eval"):
        class_path = os.path.join(f'DATA/!FINAL/train_val', class_name)

        # Create paths for train and val directories
        for split in ["train", "val"]:
            os.makedirs(os.path.join("DATA/!TEMP/", dst_name, split, class_name), exist_ok=True)
        # Split the images into train and eval sets
        train_image, eval_image = train_test_split(os.listdir(class_path), test_size=composition[1], random_state=seed)
        
        # Copy images to the respective directories
        for image in train_image:
            shutil.copy(os.path.join(class_path, image), os.path.join("DATA/!TEMP", dst_name, "train", class_name, image))
        for image in eval_image:
            shutil.copy(os.path.join(class_path, image), os.path.join("DATA/!TEMP", dst_name, "val", class_name, image))
        
        # Create a DataFrame with the number of images in each class for train and eval sets
        temp[class_name] = {
            "train": len(train_image),
            "val": len(eval_image)
        }
    df = pd.DataFrame(temp).T
    df.index.name = "Class"
    df.reset_index(inplace=True)
    df['train'] = df['train'].astype(int)
    df['val'] = df['val'].astype(int)
    return df
    
def adjustClassBalance(dir_name, target_sample_class = 100, custom = False) -> pd.DataFrame:
    '''
    Adjusts the class balance issue in the dataset located in `DATA/!TEMP/<dir>/train`.

    If classes have fewer samples than target_sample_class, they will be oversampled by augmentation.
    
    If classes have more samples than target_sample_class, they will be undersampled by random selection.

    Parameters
    ------------
    - dir: 
        The directory name where the dataset is located.
    - target_sample_class: 
        The target number of samples per class.
    - custom:
        If True, the function will use the directory name as is. If False, it will use the default path `DATA/!TEMP/<dir>/train`.
    '''
    if custom:
        dir_path = dir_name
    else:
        dir_path = os.path.join("DATA/!TEMP", dir_name, "train")
    class_names = os.listdir(dir_path)
    # Count the number of images in each class before adjustment
    old_len_classes = count_class(dir_path, returnType="count")

    for class_name in tqdm(class_names , desc="Adjusting class balance"):
        class_path = os.path.join(dir_path, class_name)
        images = os.listdir(class_path)
        current_len = len(images)

        # Oversampling if the class has fewer samples than target_sample_class
        if current_len < target_sample_class:
            index = 0
            while len(os.listdir(class_path)) < target_sample_class:
                image = rd.choice(images)
                image = Image.open(os.path.join(class_path, image))
                # random rotation
                rotation_angle = rd.randint(-20, 20) 
                image = image.rotate(rotation_angle)
                # Random horizontal flip
                if rd.random() > 0.5:
                    image = ImageOps.mirror(image)
                # zooming in or out
                if rd.random() > 0.5:
                    width, height = image.size
                    scale_factor = rd.uniform(0.8, 1.2)
                    new_width = int(width * scale_factor)
                    new_height = int(height * scale_factor)
                    image = image.resize((new_width, new_height))
                    image = image.crop((0, 0, width, height)) 
                # random brightness adjustment
                enhancer = ImageEnhance.Brightness(image)
                factor = rd.uniform(0.7, 1.3)
                image = enhancer.enhance(factor)

                # Save the augmented image
                saveName = f"augmented_{index}.jpg"
                image.save(os.path.join(class_path, saveName))
                index += 1
        
        # Undersampling if the class has more samples than target_sample_class 
        elif current_len > target_sample_class:
            selected_images = rd.sample(images, target_sample_class)
            for image in images:
                if image not in selected_images:
                    os.remove(os.path.join(class_path, image))
    
    new_len_classes = count_class(dir_path, returnType="count")

    return pd.DataFrame({
        "Class": class_names,
        "Old train Len": old_len_classes,
        "New train Len": new_len_classes
    })

def load_data_test(is_HOG = False, orientation = "", pixels_per_cell = "", cell_per_block = ""):
    """
    Loads the test dataset from the directory `DATA/!FINAL/test`.
    If `is_HOG` is True, it extracts HOG features from the images.
    If `is_HOG` is False, it flattens the images into 1D arrays.

    Parameters
    ------------
    - is_HOG (bool):
        If True, extracts HOG features from the images.
    - orientation (int):
        The number of orientation bins for HOG features. Required if `is_HOG` is True.
    - pixels_per_cell (tuple):
        The size of each cell in pixels for HOG features. Required if `is_HOG` is True.
    - cell_per_block (tuple):
        The number of cells per block for HOG features. Required if `is_HOG` is True.
    
    Returns
    ------------
    - test_X (list):
        A list of test images, either as HOG features or flattened arrays.
    - test_y (list):
        A list of labels corresponding to the test images.
    - label_names (list):
        A list of class names corresponding to the labels.
    """
    # Check if HOG parameters are provided when is_HOG is True
    if is_HOG and (not orientation or not pixels_per_cell or not cell_per_block):
        raise TimeoutError("Please put orientation, pixels_per_cell, and cell_per_block if is_HOG ON")

    DIR_PATH = os.path.join("DATA/!FINAL/test")
    label_names = os.listdir(DIR_PATH)

    test_X = []
    test_y = []
    
    if is_HOG:
        for class_dir in os.listdir(DIR_PATH):
            class_path = os.path.join(DIR_PATH, class_dir)
            for image_file in os.listdir(class_path):
                image_path = os.path.join(class_path, image_file)
                image = Image.open(image_path).convert("L")  # Convert to grayscale
                image = image.resize((224, 224))
                image_array = np.array(image) / 255.0
                hog_features = hog(
                    image_array,
                    orientations= orientation,
                    pixels_per_cell=pixels_per_cell,
                    cells_per_block=cell_per_block,
                    transform_sqrt=True,
                    block_norm='L2-Hys'
                )
                test_X.append(hog_features)
                test_y.append(label_names.index(class_dir))
    else:
        for class_dir in os.listdir(DIR_PATH):
            class_path = os.path.join(DIR_PATH, class_dir)
            for image_file in os.listdir(class_path):
                image_path = os.path.join(class_path, image_file)
                image = Image.open(image_path).convert("RGB")
                image = image.resize((224, 224))
                image_array = np.array(image) / 255.0
                test_X.append(image_array.flatten())
                test_y.append(label_names.index(class_dir))

    return test_X, test_y, label_names

def load_data_train_eval(DIR_NAME: str):
    """
    Loads the training and evaluation datasets from the directory `DATA/!TEMP/<DIR_NAME>`.
    The directory should contain subdirectories `train` and `val`, each containing class subdirectories with images.

    Parameters
    ------------
    - DIR_NAME (str):
        The name of the directory containing the training and evaluation datasets. MUST BE IN `DATA/!TEMP/`.
    """
    DIR_PATH = os.path.join("DATA/!TEMP", DIR_NAME)
    label_names = os.listdir(os.path.join(DIR_PATH, "train"))
    train_X = []
    train_y = []
    test_X = []
    test_y = []

    for data_split_type in os.listdir(DIR_PATH):
        current_path = os.path.join(DIR_PATH, data_split_type)
        if data_split_type == "train":
            for class_dir in os.listdir(current_path):
                class_path = os.path.join(current_path, class_dir)
                for image_file in os.listdir(class_path):
                    image_path = os.path.join(class_path, image_file)
                    image = Image.open(image_path).convert("RGB")
                    image = image.resize((224, 224))
                    image_array = np.array(image) / 255.0
                    train_X.append(image_array.flatten())
                    train_y.append(label_names.index(class_dir))
        elif data_split_type == "val":
            for class_dir in os.listdir(current_path):
                class_path = os.path.join(current_path, class_dir)
                for image_file in os.listdir(class_path):
                    image_path = os.path.join(class_path, image_file)
                    image = Image.open(image_path).convert("RGB")
                    image = image.resize((224, 224))
                    image_array = np.array(image) / 255.0
                    test_X.append(image_array.flatten())
                    test_y.append(label_names.index(class_dir))
                    
    return train_X, train_y, test_X, test_y

def get_metrics_SVM(model, X, y_true, class_labels, plot_confusion_matrix: Literal[True, False] = True, mode : Literal["simple", "detailed"] = "simple"):
    """
    Computes and returns various classification metrics for the SVM model.
    The metrics include True Positives, False Positives, False Negatives, Precision, Recall, F1 Score, and Balanced Accuracy.

    If `plot_confusion_matrix` is True, it plots the confusion matrix.

    If `mode` is "detailed", it returns a DataFrame with class-specific metrics and overall metrics.
    
    If `mode` is "simple", it returns a dictionary with overall metrics only.

    Parameters
    ------------
    - model: 
        The trained SVM model to evaluate.
    - X:
        The input features for the evaluation dataset.
    - y_true:
        The true labels for the evaluation dataset.
    - class_labels:
        A list of class names corresponding to the labels.
    - plot_confusion_matrix (bool):
        If True, plots the confusion matrix.
    - mode (str):
        The mode of operation. Can be "simple" or "detailed". 
        "simple" returns overall metrics only, while "detailed" returns class-specific metrics as well.
    
    Returns
    ------------
    - result (dict):
        A dictionary containing the overall metrics or a detailed report based on the `mode` parameter.
    """
    y_pred = model.predict(X).astype(int)
    y_true = np.array(y_true)

    returnClassmetrics = {}
    for index, class_name in enumerate(class_labels):
        # Get true positives, false positives, and false negatives
        tp = sum((y_true == index) & (y_pred == index))
        fp = sum((y_true != index) & (y_pred == index))
        fn = sum((y_true == index) & (y_pred != index))

        class_precision = tp / (tp + fp) if (tp + fp) > 0 else 0    # Precision = TP / (TP + FP)
        class_recall = tp / (tp + fn) if (tp + fn) > 0 else 0       # Recall = TP / (TP + FN)
        class_f1 = 2 * class_precision * class_recall / (class_precision + class_recall) if (class_precision + class_recall) > 0 else 0 # F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

        returnClassmetrics[class_name] = {
            'True Positives': int(tp),
            "False Positives": int(fp),
            "False Negatives": int(fn),
            'Precision': round(class_precision,4),
            'Recall': round(class_recall, 4),
            'F1 Score': round(class_f1,4)
        }
    
    # Calculate overall metrics (macro averaged)
    balanced_accuracy = balanced_accuracy_score(y_true, y_pred)
    overal_precision = np.mean([metrics['Precision'] for metrics in returnClassmetrics.values()])
    overall_recall = np.mean([metrics['Recall'] for metrics in returnClassmetrics.values()])
    overall_f1_score = np.mean([metrics['F1 Score'] for metrics in returnClassmetrics.values()])

    if plot_confusion_matrix:
        cm = confusion_matrix(y_true, y_pred)
        # Plot confusion matrix
        plt.figure(figsize=(4, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=class_labels, yticklabels=class_labels)
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.title('Confusion Matrix')
        plt.tight_layout()
        plt.show()
        
    if mode == "detailed":
        result = { 
            "Balanced Accuracy": round(balanced_accuracy, 4),
            "Precision": round(overal_precision, 4), 
            "Recall": round(overall_recall, 4), 
            "F1 Score": round(overall_f1_score, 4), 
            "Class Metrics": returnClassmetrics
        } 
        # Create a main DataFrame with overall metrics
        overall_metrics = {k: [v] for k, v in result.items() if k != "Class Metrics"}
        overall_df = pd.DataFrame(overall_metrics)
        print("Overall Metrics:")
        display(overall_df)

        # Create a DataFrame for class-specific metrics
        class_metrics = result["Class Metrics"]
        class_df = pd.DataFrame.from_dict(class_metrics, orient='index').sort_index()
        print("\nClass-Specific Metrics:")
        display(class_df)
        
    if mode == "simple":
        result = { 
            "Balanced Accuracy": round(balanced_accuracy, 4),
            "Precision": round(overal_precision, 4), 
            "Recall": round(overall_recall, 4), 
            "F1 Score": round(overall_f1_score, 4), 
        } 
    return result

# # NO LONGER IN USE, ONLY FOR REFERENCE and TESTING
# def simplePipeline_trainSVC(DIR_NAME: str, train_test_rate = [0.8, 0.2], target_sample_class: int = 100):
#     """
#     """
#     split_train_eval(DIR_NAME, composition=train_test_rate)
#     adjustClassBalance(DIR_NAME, target_sample_class)
    
#     # Load the data
#     test_X, test_y, class_labels = load_data_test()
#     train_X, train_y, eval_X, eval_y = load_data_train_eval(DIR_NAME)
#     wandb.init(project="OPTUNA_SVM", entity="rheyhanfahry", name=DIR_NAME)
    
#     st = time.time()
#     model = SVC()
#     model.fit(train_X, train_y)
#     et = time.time()
#     fitting_time = et-st
    
#     # log to WANDB
#     st = time.time()
#     result = get_metrics_SVM(model, eval_X, eval_y, class_labels, plot_confusion_matrix=False, mode="detailed")
#     et = time.time()
#     predict_time = et-st
    
#     # each class metrics
#     log_class = pd.DataFrame(result["Class Metrics"]).T
#     log_class[["True Positives", "False Positives", "False Negatives"]] = log_class[["True Positives", "False Positives", "False Negatives"]].astype(int)
#     log_class["Class"] = log_class.index
#     log_class = log_class[["Class", "True Positives", "False Positives", "False Negatives", "Precision", "Recall", "F1 Score"]]

#     class_table = wandb.Table(dataframe=log_class)
#     wandb.log({"each class metrics": class_table})

#     # overall metrics
#     result.pop("Class Metrics")
#     log_overall = pd.DataFrame(result, index=[0])
#     log_overall = log_overall[["Balanced Accuracy", "Precision", "Recall", "F1 Score"]]
#     overall_table = wandb.Table(dataframe=log_overall)
#     wandb.log({"overall metrics": overall_table})

#     # Process time
#     temp = pd.DataFrame([fitting_time, predict_time]).T
#     temp.columns = ["Fitting Time", "predict Time"]
#     process_time_table = wandb.Table(dataframe=temp)
#     wandb.log({"Process time": process_time_table})
    
#     shutil.rmtree("DATA/!TEMP")
#     shutil.rmtree("wandb")
#     wandb.finish()


class getBestHyperparamms_SVM:
    '''
    A class to perform hyperparameter tuning for SVM using Optuna.
    It performs stratified k-fold cross-validation, evaluates the model, and saves the results to a specified directory.

    The class also handles autosaving of results and uploading them to Google Cloud Storage.

    The hyperparameters tuned include:
    - C: Regularization parameter
    - kernel: Kernel type (linear, rbf, polynomial)
    - gamma: Kernel coefficient (for rbf and polynomial kernels)
    - degree: Degree of the polynomial kernel (if kernel is polynomial)
    - num_of_balancing: Number of balancing samples to use for class balancing

    PIPELINE:
    - Initialize the class with the number of folds and the directory to autosave.
    - Call the `start` function to begin the hyperparameter tuning process.
        - Start the tuning process by calling the `objective` function, which is used by Optuna to evaluate different hyperparameter sets and use the average f1 score of each fold to guide the search.
            - Start k_fold cross-validation using the `stratified_k_fold_cross_validation` function.
                - The `stratified_k_fold_split` function is used to split the dataset into training and validation sets.
                - The `adjustClassBalance` function is used to balance the classes in the training set.
                - The training and evaluation data are loaded for the current fold.
                - The model is trained. Then evaluated with the validation set, and the results are logged.
                - The `autosave` function is called to save the current state of the overall metrics DataFrame to a CSV file in the specified directory and upload it to Google Cloud Storage.
                - The F1 score of the current fold is returned as the objective value for Optuna to optimize.
            - Average the f1 scores of the folds and send it back to optuna to continue the search.
    - After all trials are completed, the overall metrics are saved to a CSV file and uploaded to Google Cloud Storage.
    - The average metrics of each fold within the same hyperparameter sets are saved to a separate CSV file and uploaded to Google Cloud Storage.
    '''
    def __init__(self, k:int=4, savedir: str = ""):
        self.k = k

        # LOGS FOR EACH HYPERPARAMETER SETS and its METRICS RESULT
        self.overallMetrics = pd.DataFrame(columns=["fold" , "C", "kernel", "gamma", "degree",
                                                    "num_of_balancing", "Balanced Accuracy", "Precision", "Recall", "F1 Score"])
        
        # GET LABEL_NAMES
        _, _, self.label_names = load_data_test()

        # SETUP FOR AUTOSAVING
        self.savedir = savedir
        self.index = 0
        if os .path.exists(self.savedir):
            raise FileExistsError("Please provide a new directory name, the directory already exists.")
        os.makedirs(self.savedir)

    def autosave(self):
        '''
        Better be safe than sorry.

        This method saves the current state of the overall metrics DataFrame to a CSV file in the specified directory.
        It also uploads the CSV file to Google Cloud Storage and increments the index for the next save
        '''
        local_save_path = os.path.join(self.savedir, f'{self.index}.csv')
        self.overallMetrics.to_csv(local_save_path, index=False)
        upload_to_bucket(local_save_path, local_save_path)
        upload_to_bucket("optuna_study_SVM.db", "optuna_study_SVM.db")
        self.index +=1
        
    def start(self, trial: int = 100):
        """
        Starts the hyperparameter tuning process using Optuna.

        This method checks if the Optuna study database exists in Google Cloud Storage.
        If it does, it downloads the database to the local directory.
        It then recontinues the Optuna study and starts the optimization process for the specified number of trials

        Parameters
        ------------
        - trial (int):
            The number of trials to run for hyperparameter tuning. Default is 100.
        """
        # Check if the Optuna study database exists in Google Cloud Storage 
        if "optuna_study_SVM.db" in get_all_blobs():
            bucket_to_local("optuna_study_SVM.db", "optuna_study_SVM.db")
        # Initialized Optuna
        storage = optuna.storages.RDBStorage(url="sqlite:///optuna_study_SVM.db")
        study = optuna.create_study(direction='maximize', study_name='SVM_Optuna', storage=storage, load_if_exists=True)

        # Start the hyperparameter tuning process
        study.optimize(self.objective, n_trials=trial)
        
        # Save the overall metrics to a CSV file
        self.overallMetrics.to_csv("OVERALL_SVM_TUNING.csv", index=False)
        upload_to_bucket("OVERALL_SVM_TUNING.csv", "OVERALL_SVM_TUNING.csv")

        # Save the average metrics of each fold within the same hyperparameter sets to a CSV file
        filters = [i for i in list(self.overallMetrics.columns) if i not in ["fold", "Balanced Accuracy", "Precision", "Recall", "F1 Score"]]
        self.overallMetrics[filters] = self.overallMetrics[filters].astype(str)
        avg_metrics = self.overallMetrics.groupby(filters).mean().reset_index().sort_values(by="F1 Score", ascending=False).drop(columns=["fold"])
        avg_metrics["Balanced Accuracy"] = avg_metrics["Balanced Accuracy"].round(4)
        avg_metrics["Precision"] = avg_metrics["Precision"].round(4)
        avg_metrics["Recall"] = avg_metrics["Recall"].round(4)
        avg_metrics["F1 Score"] = avg_metrics["F1 Score"].round(4)
        avg_metrics.to_csv("AVERAGE_SVM_TUNING.csv", index=False)
        upload_to_bucket("AVERAGE_SVM_TUNING.csv", "AVERAGE_SVM_TUNING.csv")

    def objective(self, trial):
        '''
        The objective function for Optuna to evaluate different hyperparameter sets.
        This function sets up the hyperparameters for the SVM model, performs stratified k-fold cross-validation,
        and evaluates the model's performance using its average F1 score as the metric.

        Parameters
        ------------
        - trial:
            The Optuna trial object that contains the hyperparameters to be evaluated.
        '''
        self.C = trial.suggest_float('C', 1e-3, 1e3, log=True)
        self.kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'polynomial'])
        self.gamma = trial.suggest_float('gamma', 1e-5, 1e-1, log=True) if self.kernel in ['rbf', 'polynomial'] else "auto"
        self.degree = trial.suggest_int('degree', 2, 4) if self.kernel == 'polynomial' else None
        self.num_of_balancing = trial.suggest_int('num_of_balancing', 250, 500, step=25)
        
        # name of the dataset with this hyperparameter set. Will be used for saving the results and such
        self.dataset_name = f"{self.C}, {self.kernel}, {self.gamma}, {self.degree}, {self.num_of_balancing}"

        # Start the stratified k-fold cross-validation and get the average F1 score
        avg_f1_val = self.stratified_k_fold_cross_validation()

        # Target for optuna to get the next set of hyperparameter set with the respect of measuring the average F1 score
        trial.set_user_attr('avg_f1_val', avg_f1_val)
        
        return avg_f1_val

    def stratified_k_fold_split(self, dst_name, composition = [0.85, 0.15], seed=rd.randint(0, 100000)):
        """
        Splits the dataset into training and validation sets while maintaining class balance.
        The dataset is expected to be organized in a directory structure where each class has its own subdirectory.

        The split is done based on the specified composition, which should sum to 1.

        The split will create a new directory structure in `DATA/!TEMP/<dst_name>/train` and `DATA/!TEMP/<dst_name>/val`.

        The function also maintains a blacklist of images that have already been used in previous splits to avoid duplication.

        Parameters
        ------------
        - dst_name (str):
            The name of the destination directory where the split datasets will be saved.
        - composition (list):
            A list containing two float values that represent the proportion of the training and validation sets, respectively.
        - seed (int):
            The random seed for reproducibility. Default is a random integer between 0 and 100000.
        """
        assert sum(composition) == 1, "Composition must sum to 1"
        
        # Split dataset into train and val equally on each class
        for class_name in tqdm(os.listdir(f'DATA/!FINAL/train_val'), desc="Splitting data into train and eval"):
            class_path = os.path.join(f'DATA/!FINAL/train_val', class_name)

            # Create paths for train and val directories
            for split in ["train", "val"]:
                os.makedirs(os.path.join("DATA/!TEMP/", dst_name, split, class_name), exist_ok=True)

            # Split the images into train and eval sets
            eval_target = int(len(os.listdir(class_path)) * composition[1])
            list_of_images = os.listdir(class_path)

            # Shuffle the images and get unique image for evaluation each fold
            rd.seed(seed)
            rd.shuffle(list_of_images)
            eval_image = [i for i in list_of_images if i not in self.blacklist[class_name]][:eval_target]
            train_image = [i for i in list_of_images if i not in eval_image]

            # Copy images to the respective directories
            for image in train_image:
                shutil.copy(os.path.join(class_path, image), os.path.join("DATA/!TEMP", dst_name, "train", class_name, image))
            for image in eval_image:
                shutil.copy(os.path.join(class_path, image), os.path.join("DATA/!TEMP", dst_name, "val", class_name, image))

            # Update the blacklist with the images used in this fold SO THEY WON'T BE USED IN THE NEXT FOLD
            self.blacklist[class_name].extend(eval_image)

    def stratified_k_fold_cross_validation(self):
        '''
        Performs stratified k-fold cross-validation on the dataset.

        This method creates a new dataset for each fold, performs class balancing, trains the SVM model,
        evaluates the model, and logs the results to WANDB.
        '''

        # Timeout handler to raise an error if training exceeds 15 minutes
        def timeout_handler():
            raise TimeoutError("SVM training exceeded 15 minutes")
        
        # For keeping track of images used in previous folds
        self.blacklist = {"Darna_trima": [], "Parasa_lepida": [], "Setora_nitens": [], "Setothosea_asigna": []}
        f1_fold = []
    
        for i in tqdm(range(self.k), desc="K-Fold Cross Validation"):
            
            # Create a new dataset for each fold
            fold_name = f"{i}_fold_{self.dataset_name}"
            dataset_path = os.path.join("DATA/!TEMP", fold_name)

            # create the split of the dataset 
            self.stratified_k_fold_split(fold_name, composition=[0.85, 0.15])
            # Adjust class balance in the dataset
            adjustClassBalance(fold_name, target_sample_class=self.num_of_balancing)
            # Load the training and evaluation data for the current fold on notebook
            train_X, train_y, eval_X, eval_y = load_data_train_eval(fold_name)

            model = SVC(C=self.C, kernel=self.kernel, gamma=self.gamma, degree = self.degree, class_weight="balanced")

            # Initialize WANDB
            wandb.init(project="OPTUNA_SVM", entity="rheyhanfahry", name=fold_name)

            # Starts training, if training exceeds 15 minutes, it will raise a TimeoutError
            st = time.time()
            timer = threading.Timer(900, timeout_handler)
            timer.start()
            try:
                model.fit(train_X, train_y)
            finally:
                timer.cancel()
            et = time.time()
            fitting_time = et - st

            # Get the metrics for the current fold with val dataset
            st = time.time()
            result = get_metrics_SVM(model, eval_X, eval_y, self.label_names, plot_confusion_matrix=False, mode="detailed")
            et = time.time()
            predict_time = et-st

            # each class metrics
            log_class = pd.DataFrame(result["Class Metrics"]).T
            log_class[["True Positives", "False Positives", "False Negatives"]] = log_class[["True Positives", "False Positives", "False Negatives"]].astype(int)
            log_class["Class"] = log_class.index
            log_class = log_class[["Class", "True Positives", "False Positives", "False Negatives", "Precision", "Recall", "F1 Score"]]
            class_table = wandb.Table(dataframe=log_class)
            wandb.log({"each class metrics": class_table})
    
            # overall metrics
            result.pop("Class Metrics")
            log_overall = pd.DataFrame(result, index=[0])
            log_overall = log_overall[["Balanced Accuracy", "Precision", "Recall", "F1 Score"]]
            overall_table = wandb.Table(dataframe=log_overall)
            wandb.log({"overall metrics": overall_table})
    
            # Process time
            temp = pd.DataFrame([fitting_time, predict_time]).T
            temp.columns = ["Fitting Time", "predict Time"]
            process_time_table = wandb.Table(dataframe=temp)
            wandb.log({"Process time": process_time_table})
            
            # Clean up the dataset path after training
            shutil.rmtree(dataset_path, ignore_errors=True)

            # Stop WANDB
            wandb.finish()
            
            # used_vram, total_vram = get_vram_usage()
            del model
            # if (used_vram / total_vram) > 0.7:
            #     gc.collect()
            #     keras.backend.clear_session()

            # Log the overall metrics for the current fold
            log_overall["C"] = self.C
            log_overall["kernel"] = self.kernel
            log_overall["gamma"] = self.gamma
            log_overall["degree"] = self.degree
            log_overall["num_of_balancing"] = self.num_of_balancing
            log_overall["fold"] = i + 1
            self.overallMetrics = pd.concat([self.overallMetrics, log_overall], ignore_index=True)
            
            clear_output(wait=True)

            # Append the F1 Score of the current fold to the list
            f1_fold.append(log_overall["F1 Score"].values[0])
            
            self.autosave()
            
        return np.mean(f1_fold)

class getBestHyperparamms_SVM_HOG:
    '''
    Almost the same as `getBestHyperparamms_SVM` but with HOG features.

    This class performs hyperparameter tuning for SVM using HOG features extracted from images.
    It performs stratified k-fold cross-validation, evaluates the model, and saves the results to a specified directory.
    The class also handles autosaving of results and uploading them to Google Cloud Storage.

    The hyperparameters tuned include:
    - C: Regularization parameter
    - kernel: Kernel type (linear, rbf, polynomial)
    - gamma: Kernel coefficient (for rbf and polynomial kernels)
    - degree: Degree of the polynomial kernel (if kernel is polynomial)
    - num_of_balancing: Number of balancing samples to use for class balancing
    - orientations: Number of orientation bins for HOG features
    - pixels_per_cell: Size of each cell in pixels for HOG features
    - cells_per_block: Number of cells per block for HOG features
    '''
    def __init__(self, k:int=4, savedir: str = ""):
        self.k = k

        # LOGS FOR EACH HYPERPARAMETER SETS and its METRICS RESULT
        self.overallMetrics = pd.DataFrame(columns=["fold" , "C", "kernel", "gamma", "degree", "orientation", "pixels_per_cell", "cell_per_block",
                                                    "num_of_balancing", "Balanced Accuracy", "Precision", "Recall", "F1 Score"])
        self.test_X, self.test_y, self.label_names = load_data_test()
        self.error = []
        
        self.savedir = savedir
        self.index = 0
        if os .path.exists(self.savedir):
            raise FileExistsError("Please provide a new directory name, the directory already exists.")
        
        os.makedirs(self.savedir)

    def autosave(self):
        '''
        Better be safe than sorry.
        '''
        local_save_path = os.path.join(self.savedir, f'{self.index}.csv')
        self.overallMetrics.to_csv(local_save_path, index=False)
        upload_to_bucket(local_save_path, local_save_path)
        upload_to_bucket("optuna_study_SVM_HOG.db", "optuna_study_SVM_HOG.db")
        self.index +=1
        
    def start(self, trial: int = 100):
        """
        """
        if "optuna_study_SVM_HOG.db" in get_all_blobs():
            bucket_to_local("optuna_study_SVM_HOG.db", "optuna_study_SVM_HOG.db")
    
        # Initialized Optuna
        storage = optuna.storages.RDBStorage(url="sqlite:///optuna_study_SVM_HOG.db")
        study = optuna.create_study(direction='maximize', study_name='SVM_HOG_Optuna', storage=storage, load_if_exists=True)
        study.optimize(self.objective, n_trials=trial)
        
        # Save the overall metrics to a CSV file
        self.overallMetrics.to_csv("OVERALL_SVM_HOG_TUNING.csv", index=False)
        upload_to_bucket("OVERALL_SVM_HOG_TUNING.csv", "OVERALL_SVM_HOG_TUNING.csv")

        filters = [i for i in list(self.overallMetrics.columns) if i not in ["fold", "Balanced Accuracy", "Precision", "Recall", "F1 Score"]]
        self.overallMetrics[filters] = self.overallMetrics[filters].astype(str)
        avg_metrics = self.overallMetrics.groupby(filters).mean().reset_index().sort_values(by="F1 Score", ascending=False).drop(columns=["fold"])
        
        avg_metrics["Balanced Accuracy"] = avg_metrics["Balanced Accuracy"].round(4)
        avg_metrics["Precision"] = avg_metrics["Precision"].round(4)
        avg_metrics["Recall"] = avg_metrics["Recall"].round(4)
        avg_metrics["F1 Score"] = avg_metrics["F1 Score"].round(4)

        avg_metrics.to_csv("AVERAGE_SVM_HOG_TUNING.csv", index=False)
        upload_to_bucket("AVERAGE_SVM_HOG_TUNING.csv", "AVERAGE_SVM_HOG_TUNING.csv")

    def objective(self, trial):
        '''
        '''
        self.C = trial.suggest_float('C', 1e-3, 1e3, log=True)
        self.kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'polynomial'])
        self.gamma = trial.suggest_float('gamma', 1e-5, 1e-1, log=True) if self.kernel in ['rbf', 'polynomial'] else "auto"
        self.degree = trial.suggest_int('degree', 2, 4) if self.kernel == 'polynomial' else None
        self.num_of_balancing = trial.suggest_int('num_of_balancing', 250, 500, step=25)

        self.orientations = trial.suggest_int('orientations', 6, 10)
        self.pixels_per_cell = trial.suggest_categorical('pixels_per_cell', [(12, 12), (16, 16), (24, 24)])
        self.cells_per_block = trial.suggest_categorical('cells_per_block', [(2, 2), (3, 3)])

        self.dataset_name = f"{self.C}, {self.kernel}, {self.gamma}, {self.degree}, {self.num_of_balancing}, {self.orientations}, {self.pixels_per_cell}, {self.cells_per_block}"
        avg_f1_val = self.stratified_k_fold_cross_validation()
        trial.set_user_attr('avg_f1_val', avg_f1_val)

        return avg_f1_val
    
    def stratified_k_fold_split(self, dst_name, composition = [0.85, 0.15], seed=rd.randint(0, 100000)):
        """
        """
        assert sum(composition) == 1, "Composition must sum to 1"
        
        # Split dataset into train and val equally on each class
        for class_name in tqdm(os.listdir(f'DATA/!FINAL/train_val'), desc="Splitting data into train and eval"):
            class_path = os.path.join(f'DATA/!FINAL/train_val', class_name)

            # Create paths for train and val directories
            for split in ["train", "val"]:
                os.makedirs(os.path.join("DATA/!TEMP/", dst_name, split, class_name), exist_ok=True)

            # Split the images into train and eval sets
            eval_target = int(len(os.listdir(class_path)) * composition[1])
            list_of_images = os.listdir(class_path)

            # Shuffle the images and get unique image for evaluation each fold
            rd.seed(seed)
            rd.shuffle(list_of_images)
            eval_image = [i for i in list_of_images if i not in self.blacklist[class_name]][:eval_target]
            train_image = [i for i in list_of_images if i not in eval_image]


            # Copy images to the respective directories
            for image in train_image:
                shutil.copy(os.path.join(class_path, image), os.path.join("DATA/!TEMP", dst_name, "train", class_name, image))
            for image in eval_image:
                shutil.copy(os.path.join(class_path, image), os.path.join("DATA/!TEMP", dst_name, "val", class_name, image))
            
            self.blacklist[class_name].extend(eval_image)

    def load_data_train_HOG(self, DIR_NAME):
        DIR_PATH = os.path.join("DATA/!TEMP", DIR_NAME)
        label_names = os.listdir(os.path.join(DIR_PATH, "train"))
        train_X = []
        train_y = []
        eval_X = []
        eval_y = []

        for data_split_type in os.listdir(DIR_PATH):
            current_path = os.path.join(DIR_PATH, data_split_type)
            if data_split_type == "train":
                for class_dir in os.listdir(current_path):
                    class_path = os.path.join(current_path, class_dir)
                    for image_file in os.listdir(class_path):
                        image_path = os.path.join(class_path, image_file)
                        image = Image.open(image_path).convert("L")  # Convert to grayscale
                        image = image.resize((224, 224))
                        image_array = np.array(image) / 255.0
                        hog_features = hog(
                            image_array,
                            orientations=self.orientations,
                            pixels_per_cell=self.pixels_per_cell,
                            cells_per_block=self.cells_per_block,
                            transform_sqrt=True,
                            block_norm='L2-Hys'
                        )
                        train_X.append(hog_features)
                        train_y.append(label_names.index(class_dir))
            elif data_split_type == "val":
                for class_dir in os.listdir(current_path):
                    class_path = os.path.join(current_path, class_dir)
                    for image_file in os.listdir(class_path):
                        image_path = os.path.join(class_path, image_file)
                        image = Image.open(image_path).convert("L")  # Convert to grayscale
                        image = image.resize((224, 224))
                        image_array = np.array(image) / 255.0
                        hog_features = hog(
                            image_array,
                            orientations=self.orientations,
                            pixels_per_cell=self.pixels_per_cell,
                            cells_per_block=self.cells_per_block,
                            transform_sqrt=True,
                            block_norm='L2-Hys'
                        )
                        eval_X.append(hog_features)
                        eval_y.append(label_names.index(class_dir))
                        
        return train_X, train_y, eval_X, eval_y


    def stratified_k_fold_cross_validation(self):
        '''
        '''
        def timeout_handler():
            raise TimeoutError("SVM training exceeded 15 minutes")
        
        self.blacklist = {"Darna_trima": [], "Parasa_lepida": [], "Setora_nitens": [], "Setothosea_asigna": []}
        f1_fold = []
    
        for i in tqdm(range(self.k), desc="K-Fold Cross Validation"):
            
            # Create a new dataset for each fold
            fold_name = f"{i}_fold_{self.dataset_name}"
            dataset_path = os.path.join("DATA/!TEMP", fold_name)

            self.stratified_k_fold_split(fold_name, composition=[0.85, 0.15])
            adjustClassBalance(fold_name, target_sample_class=self.num_of_balancing)

            train_X, train_y, eval_X, eval_y = self.load_data_train_HOG(fold_name)
            
            model = SVC(C=self.C, kernel=self.kernel, gamma=self.gamma, degree = self.degree, class_weight="balanced")

            # Initialize WANDB
            wandb.init(project="OPTUNA_SVM_HOG", entity="rheyhanfahry", name=fold_name)
            
            st = time.time()
            timer = threading.Timer(900, timeout_handler)
            timer.start()
            try:
                model.fit(train_X, train_y)
            finally:
                timer.cancel()
            et = time.time()
            fitting_time = et - st
    
            st = time.time()
            result = get_metrics_SVM(model, eval_X, eval_y, self.label_names, plot_confusion_matrix=False, mode="detailed")
            et = time.time()
            predict_time = et-st

            # each class metrics
            log_class = pd.DataFrame(result["Class Metrics"]).T
            log_class[["True Positives", "False Positives", "False Negatives"]] = log_class[["True Positives", "False Positives", "False Negatives"]].astype(int)
            log_class["Class"] = log_class.index
            log_class = log_class[["Class", "True Positives", "False Positives", "False Negatives", "Precision", "Recall", "F1 Score"]]
    
            class_table = wandb.Table(dataframe=log_class)
            wandb.log({"each class metrics": class_table})
    
            # overall metrics
            result.pop("Class Metrics")
            log_overall = pd.DataFrame(result, index=[0])
            log_overall = log_overall[["Balanced Accuracy", "Precision", "Recall", "F1 Score"]]
            overall_table = wandb.Table(dataframe=log_overall)
            wandb.log({"overall metrics": overall_table})
    
            # Process time
            temp = pd.DataFrame([fitting_time, predict_time]).T
            temp.columns = ["Fitting Time", "predict Time"]
            process_time_table = wandb.Table(dataframe=temp)
            wandb.log({"Process time": process_time_table})
            
            # Clean up the dataset path after training
            shutil.rmtree(dataset_path, ignore_errors=True)
            
            wandb.finish()
            
            # used_vram, total_vram = get_vram_usage()
            del model
            # if (used_vram / total_vram) > 0.7:
            #     gc.collect()
            #     keras.backend.clear_session()
            
            log_overall["C"] = self.C
            log_overall["kernel"] = self.kernel
            log_overall["gamma"] = self.gamma
            log_overall["degree"] = self.degree
            log_overall["num_of_balancing"] = self.num_of_balancing
            log_overall["orientation"] = self.orientations
            log_overall["pixels_per_cell"] = str(self.pixels_per_cell)
            log_overall["cell_per_block"] = str(self.cells_per_block)
            log_overall["fold"] = i + 1

            self.overallMetrics = pd.concat([self.overallMetrics, log_overall], ignore_index=True)
            
            clear_output(wait=True)
            f1_fold.append(log_overall["F1 Score"].values[0])
            
            self.autosave()
            
        return np.mean(f1_fold)

def average_Kfold_latest(DIR, csv_name = ""):
    """
    Averages the latest K-fold results from CSV files in the specified directory.
    This function reads the latest CSV file in the directory, calculates the average metrics for each unique combination of features,
    and saves the averaged results to a new CSV file.
    
    Parameters
    ------------
    - DIR (str): The directory containing the K-fold CSV files.
    - csv_name (str): The name of the output CSV file to save the averaged results
    """
    latest_csv_index = ((sorted([int(re.search(r"\d+", i)[0]) for i in os.listdir(DIR)])[-1]))
    latest_csv_path = os.path.join(DIR, f'{latest_csv_index}.csv')
    table = pd.read_csv(latest_csv_path)
    
    filters = [i for i in list(table.columns) if i not in ["fold", "Balanced Accuracy", "Precision", "Recall", "F1 Score"]]
    table[filters] = table[filters].astype(str)
    
    avg_metrics = table.groupby(filters).mean().reset_index().sort_values(by="F1 Score", ascending=False).drop(columns=["fold"])
    
    avg_metrics["Balanced Accuracy"] = avg_metrics["Balanced Accuracy"].round(4)
    avg_metrics["Precision"] = avg_metrics["Precision"].round(4)
    avg_metrics["Recall"] = avg_metrics["Recall"].round(4)
    avg_metrics["F1 Score"] = avg_metrics["F1 Score"].round(4)
    
    avg_metrics.to_csv(csv_name, index=False)    
        
def load_allData_and_balance(num_of_balancing, is_HOG = False, orientation = "", pixels_per_cell = "", cell_per_block = ""):
    """
    Loads all data from the training and validation directories, applies class balancing,
    and extracts features using HOG if specified.

    Parameters
    ------------
    - num_of_balancing (int): 
        The number of samples to balance each class to.
    - is_HOG (bool): 
        Whether to extract HOG features from the images. Default is False.
    - orientation (int): 
        The number of orientations for HOG features. Required if is_HOG is True.
    - pixels_per_cell (tuple): 
        The size of each cell in pixels for HOG features. Required if is_HOG is True.
    - cell_per_block (tuple): 
        The number of cells per block for HOG features. Required if is_HOG is True
    
    Returns
    ------------
    - train_X (list): 
        List of training features.
    - train_y (list): 
        List of training labels.
    """
    if is_HOG and (not orientation or not pixels_per_cell or not cell_per_block):
        raise TimeoutError("Please put orientation, pixels_per_cell, and cell_per_block if is_HOG ON")
    
    src_folder_path = "DATA/!FINAL/train_val"
    dst_folder_path = os.path.join("DATA", "TEMP")
    label_names = os.listdir(src_folder_path)
    shutil.copytree(src_folder_path, dst_folder_path)
    adjustClassBalance(dst_folder_path, num_of_balancing, custom=True)

    train_X = []
    train_y = []
    if is_HOG:
        for class_dir in label_names:
            class_path = os.path.join(dst_folder_path, class_dir)
            for image_file in os.listdir(class_path):
                image_path = os.path.join(class_path, image_file)
                image = Image.open(image_path).convert("L")  # Convert to grayscale
                image = image.resize((224, 224))
                image_array = np.array(image) / 255.0
                hog_features = hog(
                    image_array,
                    orientations= orientation,
                    pixels_per_cell=pixels_per_cell,
                    cells_per_block=cell_per_block,
                    transform_sqrt=True,
                    block_norm='L2-Hys'
                )
                train_X.append(hog_features)
                train_y.append(label_names.index(class_dir))
    else:
        for class_dir in label_names:
            class_path = os.path.join(dst_folder_path, class_dir)
            for image_file in os.listdir(class_path):
                image_path = os.path.join(class_path, image_file)
                image = Image.open(image_path).convert("RGB")
                image = image.resize((224, 224))
                image_array = np.array(image) / 255.0
                train_X.append(image_array.flatten())
                train_y.append(label_names.index(class_dir))
    shutil.rmtree(dst_folder_path)
    
    return train_X, train_y
```
